# Research Questions
1. How many unique shops are in the dataset

```python
uniqueShops = df['shopid'].unique()
count = (len(uniqueShops))
print(f"We have {count} unique shops in the dataset")
```

2. How many preferred and cross border shops are in the dataset

```python
preferredShops = df['is_preferred'].fillna(0).sum()
cb_option = df['cb_option'].fillna(0).sum()

print("Sum of preferred shops is:", preferredShops)
print("Sum of shops accross border is:", cb_option)
```

3. How many products have zero sold count

```python
zeroSoldCount = (df['sold_count'] == 0).sum() 
print(f"There is {zeroSoldCount} products with zero sold count")
```

4. How many products were created in the 2018

```python
# Changing 'creation date' format to date format
df['item_creation_date'] = pd.to_datetime(df['item_creation_date'])

# Calculating products created in 2018
productsCreatedIn2018 = df[df['item_creation_date'].dt.year == 2018]
countProducts2018 = len(productsCreatedIn2018)
print(f"There is {countProducts2018} products created in 2018")

```

5. Show top 3 preferred shops (shopid) that have the largest number of unique products

```python
# Filter the DataFrame for preferred shops
preferredShopsDf = df[df['is_preferred'] == True]

# Group by 'shopid' and count unique 'itemid'
uniqueProductsByshop = preferredShopsDf.groupby('shopid')['itemid'].nunique()

# Sort the groups in descending order of unique product count
topShops = uniqueProductsByshop.sort_values(ascending=False)

# Select the top 3 shops
top3Shops = topShops.head(3)

top3Shops

```

6. Show top 3 categories that have the largest number of unique cross border products

```python
# Group by 'category' and count unique 'cb_option'
uniqueCBoption = df.groupby('category')['cb_option'].nunique()

# Sort the groups in descending order of unique cross border count
topCategories = uniqueCBoption.sort_values(ascending=False)

# Select the top 3 shops
top3Categories = topCategories.head(3)

top3Categories

```

7. Find top 3 shopid with highest revenue (Assumption: the product price has not changed)

```python
# Group by 'shopid' and calculate the sum of 'sold_count'
shops = df.groupby('shopid')['sold_count'].sum()

# Sort the groups in descending order of sold_count
topShops = shops.sort_values(ascending=False)

# Select the top 3 shops
top3Shops = topShops.head(3)

top3Shops

```

8. Find number of products that have more than 3 variations (don't include products with 3 or fewer variations)

```python
# Group by 'itemid' and count variations
variationCounts = df.groupby('itemid').size()

# Filter for products with more than 3 variations
products_more_than_three_variations = variationCounts[variationCounts > 3]

# Count the number of products with more than 3 variations
numProductsMoreThanThreeVariations = len(products_more_than_three_variations)

print(f"There are {numProductsMoreThanThreeVariations} products with more than 3 variatons in the dataset")
```

9. Identify duplicated listings within each shop (If list A and B in shop S have exactly same product title, product detailed description, and price, both listing A and B are considered duplicate listings)

```python
# Identifying duplicates
df['is_duplicate'] = df.duplicated(subset=['shopid', 'item_name', 'item_description', 'price'], keep=False)

# Filter the DataFrame to show only duplicates
duplicatesDf = df[df['is_duplicate']]

#Displaying the Dataset with duplicated listings only
duplicatesDf
```

10. Mark those duplicate listings with True otherwise False and store the marking result in a new column named is_duplicated

```python
# Identifying duplicates based on the specified criteria
df['is_duplicate'] = df.duplicated(subset=['shopid', 'item_name', 'item_description', 'price'], keep=False)

#Displaying the Dataset with all the columns including the duplicated listings
df
```

11. Find duplicate listings that has less than 2 sold count and store the result in a new excel file named "duplicated_listings.xlsx

```python
# Filter duplicates with sold_count less than 2
filteredDuplicates = df[(df['is_duplicate']) & (df['sold_count'] < 2)]

# File path for the Excel file
filePath = 'duplicated_listings.xlsx'

# Check if the file already exists
if not os.path.exists(filePath):
    # Export to Excel if the file does not exist
    filteredDuplicates.to_excel(filePath, index=False)
    print(f"File '{filePath}' has been created.")
else:
    print(f"File '{filePath}' already exists. We are not overwriting.")
```
